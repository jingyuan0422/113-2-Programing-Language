```python
# Select a travel destination.
def travel_destination():
    while True:
        try:
            destination_choice = int(input("Please select a travel destination (1.台北市 2.新北市 3.基隆市): "))
            if destination_choice == 1:
                destination = "台北市"
                break
            elif destination_choice == 2:
                destination = "新北市"
                break
            elif destination_choice == 3:
                destination = "基隆市"
                break
            else:
                print("⚠️ Please enter 1, 2 or 3.")
        except ValueError:
            print("⚠️ Please enter 1, 2 or 3.")
    return destination

# Select the number of travel days
def travel_days(destination):
    while True:
        try:
            days_choice = int(input("Please enter the number of travel days (1-5 days).："))
            if 1 <= days_choice <= 2:
                days = days_choice
                break
            elif 3 <= days_choice <= 5:
                days = days_choice
                break
            else:
                print("⚠️ The number of travel days must be between 1 and 5.")
        except ValueError:
            print("⚠️ Please enter a number between 1 and 5.")
    return days

# Select a travel type.
def travel_type():
    while True:
        try:
            type_choice = int(input("Please select a travel type (1.自然 2.文化 3.休閒): "))
            if type_choice == 1:
                main_type = "自然"
                break
            elif type_choice == 2:
                main_type = "文化"
                break
            elif type_choice == 3:
                main_type = "休閒"
                break
            else:
                print("⚠️ Please enter 1, 2 or 3.")
        except ValueError:
            print("⚠️ Please enter 1, 2 or 3.")
    return main_type

# Travel preference
user_destination = travel_destination()
user_days = travel_days(user_destination)
user_type = travel_type()

print(f"✅ Your selected travel destination：{user_destination}")
if user_days >= 3:
    print(f"✅ Your selected number of travel days：{user_days} (May include areas outside of {user_destination})")
else:
    print(f"✅ Your selected number of travel days：{user_days}")
print(f"✅ Your selected travel type：{user_type}")
```

```python
prompt = f"""
我想請你根據以下資料幫我排定一個旅遊行程：
1. 景點：{', '.join(filter_df['景點'].tolist())}
2. 旅遊天數需求：{user_days}天

請根據這些景點以及旅遊天數需求安排旅遊行程。除了這些景點外，你也可以加入一些附近的景點或是當地的特色。
每一天的行程應該不會過於擁擠，且考慮到景點之間的交通時間或距離。

我需要有每天的時間、景點、縣市、交通方式、特色
請幫我寫成以下範例：
Day 1: 時間: 9:00 AM - 12:00 PM 景點: 台北 101 交通方式: 公車 特色: 擁有壯麗的觀景台，觀賞台北市全景。時間: 1:00 PM - 4:00 PM 景點: 故宮博物院交通方式: 地鐵 特色: 擁有豐富的中華文化遺產，特別是古代的藝術品。

請開始規劃，並提供每一天的詳細行程安排。
"""
```

```python
prompt
```

```python
from google.colab import userdata
apikey = userdata.get('GOOGLE_API_KEY')

import google.generativeai as genai
genai.configure(api_key = apikey)
model = genai.GenerativeModel("gemini-2.0-flash")
response = model.generate_content(prompt)
print(response.text)
```

```python
import pandas as pd

# 分割行程，每一行對應一天
lines = response.strip().split("\n")
lines = [line.replace("*", "").strip() for line in lines] # 刪除所有*

# 初始化空的列表來儲存每天的行程
days = []
times = []
sights = []
transportation = []
features = []


# 解析每一行
current_day = None
for line in lines:
    # 檢查行是否是"Day X"
    if line.startswith("Day"):
        current_day = line.strip()
    elif line.startswith("時間"):
        time = line.split(":", 1)[1].strip()  # `split(":", 1)` 只切一次，保留完整時間範圍
        times.append(time)
    elif line.startswith("景點"):
        sight = line.split(":")[1].strip()
        sights.append(sight)
    elif line.startswith("交通方式"):
        transport = line.split(":")[1].strip()
        transportation.append(transport)
    elif line.startswith("特色"):
        feature = line.split(":")[1].strip()
        features.append(feature)
        # Append the current day for each entry
        days.append(current_day)  # Moved this line here
```

```python
# 確保所有 list 長度相等（填補缺失值）
max_len = max(len(days), len(times), len(sights), len(transportation), len(features))

# 使用 `itertools.zip_longest` 確保長度一致，缺少的值填入 "N/A"
from itertools import zip_longest
data = list(zip_longest(days, times, sights, transportation, features, fillvalue="N/A"))

# 創建 DataFrame
df = pd.DataFrame(data, columns=["Day", "Time", "Sight", "Transportation", "Feature"])
```
